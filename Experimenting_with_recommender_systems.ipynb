### Hollywood movie recommendation

```python
import numpy as np
import pandas as pd
pd.set_option('display.max_columns', None)
import difflib
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

from google.colab import drive
drive.mount('/content/drive')
```

```python
movies_data = pd.read_csv('/content/drive/MyDrive/Portfolio/1. Capstone Projects/Recommender System/movies.csv')

selected_features = ['genres','keywords','tagline','cast','director']

for feature in selected_features:
    movies_data[feature] = movies_data[feature].fillna('')

combined_features = movies_data['genres']+' '+movies_data['keywords']+' '+movies_data['tagline']+' '+movies_data['cast']+' '+movies_data['director']

vectorizer = TfidfVectorizer()

feature_vectors = vectorizer.fit_transform(combined_features)

similarity = cosine_similarity(feature_vectors)

movie_name = input(' Enter your favourite movie name : ')

list_of_all_titles = movies_data['title'].tolist()

find_close_match = difflib.get_close_matches(movie_name, list_of_all_titles)

close_match = find_close_match[0]

index_of_the_movie = movies_data[movies_data.title == close_match]['index'].values[0]

similarity_score = list(enumerate(similarity[index_of_the_movie]))

sorted_similar_movies = sorted(similarity_score, key = lambda x:x[1], reverse = True)  

print('Movies suggested for you : \n')

i = 1

for movie in sorted_similar_movies:
  index = movie[0]
  title_from_index = movies_data[movies_data.index==index]['title'].values[0]
  if (i<30):
    print(i, '.',title_from_index)
    i+=1

```

### Bollywood movie recommendation

```python
import numpy as np
import pandas as pd
pd.set_option('display.max_columns', None)
import difflib
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

movies_data = pd.read_csv('/content/drive/MyDrive/Portfolio/1. Capstone Projects/Recommender System/BollywoodMovieDetail.csv')

movies_data.reset_index()

selected_features = ['title','genre','writers','actors','directors']

for feature in selected_features:
    movies_data[feature] = movies_data[feature].fillna('')

combined_features = movies_data['title']+' '+movies_data['genre']+' '+movies_data['writers']+' '+movies_data['actors']+' '+movies_data['directors']

vectorizer = TfidfVectorizer()

feature_vectors = vectorizer.fit_transform(combined_features)

similarity = cosine_similarity(feature_vectors)

movie_name = input('Enter your favourite movie name :')
list_of_all_titles = movies_data['title'].tolist()

find_close_match = difflib.get_close_matches(movie_name, list_of_all_titles)

close_match = find_close_match[0]

index_of_the_movie = movies_data[movies_data['title'] == close_match].index.values[0]

similarity_score = list(enumerate(similarity[index_of_the_movie]))

sorted_similar_movies = sorted(similarity_score, key = lambda x:x[1], reverse = True) 

print('Movies suggested for you : \n')

i = 1

for movie in sorted_similar_movies:
  index = movie[0]
  title_from_index = movies_data[movies_data.index==index]['title'].values[0]
  if (i<10):
    print(i, '.',title_from_index)
    i+=1
```

### Movie classifier model

```python
import pandas as pd
pd.set_option('display.max_columns', None)
import numpy as np
```

```python
from ast import literal_eval
df['genre'] = df['genres'].fillna('[]').apply(literal_eval).apply(lambda x: [i['name'] for i in x] if isinstance(x,list) else [])
```

```python
df['language'] = df['spoken_languages'].fillna('[]').apply(literal_eval).apply(lambda x: [i['name'] for i in x] if isinstance(x,list) else [])
```

```python
df['production_company'] = df['production_companies'].fillna('[]').apply(literal_eval).apply(lambda x: [i['name'] for i in x] if isinstance(x,list) else [])
```

```python
df['production_country'] = df['production_countries'].fillna('[]').apply(literal_eval).apply(lambda x: [i['name'] for i in x] if isinstance(x,list) else [])
```

```python
df['release_year'] = df['release_date'].apply(lambda x: str(x).split('-')[0] if x != np.nan else np.nan)
```

```python
df.head(2)
```

```python
selected_features = df[['title','genre','language','release_year','runtime','production_company','production_country','tagline','vote_average','vote_count']]
```

```python
for feature in selected_features:
    df[feature] = df[feature].fillna('')
```

```python
new_df1 = selected_features.explode('genre')
```

```python
new_df2 = new_df1.explode('language')
```

```python
new_df3 = new_df2.explode('production_company')
```

```python
final_df1 = new_df3.explode('production_country')
```

```python
final_df = final_df1.reset_index()
```

```python
final_df.drop('index', inplace=True, axis=1)
```

```python
final_df = final_df.reset_index()
```

```python
final_df
```

```python
def classified_movie(genre, percentile=0.80):
    df = final_df[final_df['genre'] == genre]
    vote_counts = df[df['vote_count'].notnull()]['vote_count'].astype('int')
    vote_averages = df[df['vote_average'].notnull()]['vote_average'].astype('int')
    C = vote_averages.mean()
    m = vote_counts.quantile(percentile)
    
    qualified = df[(df['vote_count'] >= m) & (df['vote_count'].notnull()) & (df['vote_average'].notnull())][['title','genre','language','release_year','runtime','production_company','production_country','tagline','vote_average','vote_count']]
    qualified['vote_count'] = qualified['vote_count'].astype('int')
    qualified['vote_average'] = qualified['vote_average'].astype('int')
    
    qualified['wr'] = qualified.apply(lambda x: (x['vote_count']/(x['vote_count']+m) * x['vote_average']) + (m/(m+x['vote_count']) * C), axis=1)
    qualified = qualified.sort_values('wr', ascending=False).head(250)
    
    return qualified
```

```python
#Genres
final_df['genre'].unique()
```

```python
romantice_movies = classified_movie('Science Fiction')
romantice_movies
```

```python
def get_movie_recommendation(genre, runtime, rating):
    df1 = final_df[final_df['genre'] == genre]
    df2 = df1[df1['runtime'] >= runtime]
    df3 = df2[df2['vote_average'] >= rating]
    return df3
```

```python
a = str(input('Genre:'))
b = float(input('Runtime:'))
c = float(input('Rating:'))
relevant_movies = get_movie_recommendation(a,b,c)
relevant_movies
```
