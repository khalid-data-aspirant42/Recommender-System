### Requirements (Libraries and Dataset)

```python
import numpy as np
import pandas as pd
pd.set_option('display.max_column',None)
import difflib
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import linear_kernel, cosine_similarity
from ast import literal_eval

from google.colab import drive
drive.mount('/content/drive')
```

```python
movies_data = pd.read_csv('/content/drive/MyDrive/Portfolio/1. Capstone Projects/Recommender System/prime_movies.csv')
```

```python
# features selection
selected_features = ['Language','Plot','IMDb Rating','Year of Release','Maturity Rating']

# null value treatment
for feature in selected_features:
    movies_data[feature] = movies_data[feature].fillna('')
```

```python
# combining column values into sentence
movies_data['total'] = movies_data['Language']+' '+movies_data['Plot']+' '+movies_data['IMDb Rating']+' '+movies_data['Year of Release']+' '+movies_data['Maturity Rating']
```

```python
# vector modeling
tf = TfidfVectorizer(analyzer='word',ngram_range=(1, 2),min_df=0, stop_words='english')
tfidf_matrix = tf.fit_transform(movies_data['total'])
```

```python
# finding similarity index (single movie to all movies)
cosine_sim = linear_kernel(tfidf_matrix, tfidf_matrix)
```

```python
titles = movies_data['Movie Name']
indices = pd.Series(movies_data.index, index=movies_data['Movie Name'])
```

```python
new_df = movies_data[movies_data['Language']=='Hindi']
new_df = new_df[new_df['Year of Release']>='2018'] 
new_df = new_df[new_df['IMDb Rating']>='7']
```

```python
# Creating function
def get_recommendations(title):
    idx = indices[title]
    sim_scores = list(enumerate(cosine_sim[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = sim_scores[1:31]
    movie_indices = [i[0] for i in sim_scores]
    return titles.iloc[movie_indices]

# getting inputs
get_title = str(input())
get_recommendations(get_title)
```
